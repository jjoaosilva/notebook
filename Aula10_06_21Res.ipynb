7. Faça um Programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 6 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam RS 80,00 ou em galões de 3,6 litros, que custam RS 25,00. Informe ao usuário as quantidades de tinta a serem compradas e os respectivos preços em 3 situações:
- comprar apenas latas de 18 litros;
- comprar apenas galões de 3,6 litros;
- misturar latas e galões, de forma que o
desperdício de tinta seja menor. Acrescente 10% de folga e sempre arredonde os valores para cima, isto é, considere latas cheias.


```python
import math

area = float(input("Informe o tamanho da área a ser pintada em m2: "))

litrosNecessarios = area/6 

#Apenas com latas de 18l
qtde_latas = math.ceil(litrosNecessarios/18)
preco = qtde_latas*80 

print("(A) A quantidade de latas é %d e o preço é R$%.2f \n\n" %(qtde_latas, preco))

qtde_galoes = math.ceil(litrosNecessarios/3.6)
preco = qtde_galoes*25 #

print("(B) A quantidade de galoes é %d e o preço é R$%.2f \n\n" %(qtde_galoes, preco))

litrosNecessariosComFolga = litrosNecessarios*1.1 

latas = litrosNecessariosComFolga//18
sobra = litrosNecessariosComFolga%18
galoes = math.ceil(sobra/3.6)

precoTotal = latas*80 + galoes*25
print("(C)A quantidade de latas é %d e de galões é %d e custa R$%.2f" %(latas, galoes,precoTotal))
```

16. Faça um programa que calcule as raízes de uma equação do segundo grau, na
forma ax2 + bx + c. O programa deverá pedir os valores de a, b e c e fazer as
consistências, informando ao usuário nas seguintes situações:

*   Se o usuário informar o valor de A igual a zero, a equação não é do segundo
grau e o programa não deve fazer pedir os demais valores, sendo encerrado;
*   Se o delta calculado for negativo, a equação não possui raizes reais. Informe
ao usuário e encerre o programa;
*   Se o delta calculado for igual a zero a equação possui apenas uma raiz real;
informe-a ao usuário;
*   Se o delta for positivo, a equação possui duas raiz reais; informe-as ao
usuário;

```python
import math

a = int(input('Digite a: '))
if (a == 0):
    print("A equação não é do segundo grau.")
    exit()
b = int(input('Digite b: '))
c = int(input('Digite c: '))

delta = b**2 - 4 * a * c

if (delta < 0):
    print("A equação não possui raízes reais.")
else:
    x1 = (-b + math.sqrt(delta))/2*a
    x2 = (-b - math.sqrt(delta))/2*a
    if (delta == 0):
        print("A equação possui apenas uma raiz real: ", x1)
    else:
        print("A equação possui as raízes: %f e %f." %(x1,x2))
```

17. Faça um Programa que leia um número inteiro menor que 1000 e imprima a
quantidade de centenas, dezenas e unidades do mesmo. Observando os termos no plural a colocação do "e", da vírgula entre outros. Exemplo:
326 = 3 centenas, 2 dezenas e 6 unidades
12 = 1 dezena e 2 unidades Testar com: 326, 300, 100, 320, 310,305, 301, 101, 311, 111, 25, 20, 10, 21, 11, 1, 7 e 16

```python
entradaOriginal = int(input("Insira um numero: "))

centenas = entradaOriginal//100
entradaSemCentenas = entradaOriginal%100

dezenas = entradaSemCentenas//10

unidades = entradaSemCentenas%10

stringCentena = ""
stringDezena = ""
stringUnidade = ""

if centenas > 0:
  stringCentena += str(centenas)+" centena"
  if centenas > 1:
    stringCentena += 's'

if dezenas > 0:
  stringDezena += str(dezenas)+" dezena"
  if dezenas > 1:
    stringDezena += 's'

if unidades > 0:
  stringUnidade += str(unidades)+" unidade"
  if unidades > 1:
    stringUnidade += 's'


stringFinal = stringCentena
if stringDezena != "":
  if stringFinal == "":
    stringFinal += stringDezena
  elif stringUnidade == "":
    stringFinal += " e " + stringDezena
  else:
    stringFinal += ", " + stringDezena

if stringUnidade != "":
  if stringFinal == "":
    stringFinal += stringUnidade
  else:
    stringFinal += " e " + stringUnidade

stringFinal = str(entradaOriginal) + " = " + stringFinal

print(stringFinal)
```

19. Um posto está vendendo combustíveis com a seguinte tabela de descontos:

Álcool:

*   até 20 litros, desconto de 3% por litro
*   acima de 20 litros, desconto de 5% por litro

Gasolina:

*   até 20 litros, desconto de 4% por litro
*   acima de 20 litros, desconto de 6% por litro

Escreva um algoritmo que leia o número de litros vendidos, o tipo de
combustível (codificado da seguinte forma: A-álcool, G-gasolina), calcule e
imprima o valor a ser pago pelo cliente sabendo-se que o preço do litro da
gasolina é R$ 2,50 o preço do litro do álcool é R$ 1,90.

```python
v = 0
c = input('Álcool ou gasolina?: ')
l = float(input('Quantos litros?: '))
if (c == 'A'):
    if(l <= 0):
        print("Quantia inválida")
    elif(l <= 20):
        v = 1.90*l * 97/100
    else:
        v = 1.90*l * 95/100
elif (c == 'G'):
    if(l <= 0):
        print("Quantia inválida")
    elif(l <= 20):
        v = 2.50*l * 96/100
    else:
        v = 2.50*l * 94/100
else:
    print("Tipo de combustível inválido")

print("O valor é R$%.2f" %v)
```
